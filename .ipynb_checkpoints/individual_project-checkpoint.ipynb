## (1) Data Description

The data for this project are from the UBC Computer Science research group that studies player behaviour on a customized Minecraft server. There are two datasets: players.csv and sessions.csv. Together, they record how players interact with the game and provide basic information about each player.

players.csv  
- Number of observations: 196 
- Number of variables: 7  
- Each row represents one player  
Variables include:
- experience (chr): level of previous game experience (Pro, Veteran, Amatuer, Regular or Beginner)
- subscribe (lgl): whether the player subscribed to the newsletter (TRUE or FALSE)
- hashedEmail (chr): unique player identifier that connects to sessions.csv
- played_hours (dbl): total number of hours the player has spent playing
- name (chr): player’s in-game name
- gender (chr): self-identified gender (Female, Male, Non-binary, Agender, Two-spirited, Prefer not to say or Other)
- age (dbl): age of the player in years
  
sessions.csv  
- Number of observations: 1,535  
- Number of variables: 5  
- Each row represents one play session  
Variables include:
- hashedEmail (chr): unique player identifier that connects to players.csv
- start_time (chr): when a play session began
- end_time (chr): when a play session ended
- original_start_time (dbl): unadjusted session start time
- original_end_time (dbl): unadjusted session end time

From these two datasets, I can also create new variables such as:  
- session_duration = end_time - start_time  
- sessions_per_player = number of sessions per player  
- peak_play_hour = hour of day when most sessions occur  

Issues seen in the data:  
- Missing values in gender and experience columns  
- Some inconsistencies in how timestamps are formatted (for example, different date and time styles or incomplete timestamps)  
- Possible duplicate or mismatched player IDs between datasets  
- Since the timestamp data were collected through automated server logs, it will be important to check for errors like duplicates, inconsistent IDs, or missing records that can happen in automated systems  

Other potential issues:  
- Data may be incomplete due to players disconnecting or server interruptions  
- Some variables might not be updated regularly, which could affect accuracy  

How the data were collected:  
- All information was automatically recorded by the Minecraft research server as players interacted with the game.  
- Player details such as gender, experience, and email were collected during registration, and session details were logged whenever a player joined or left the server.

## (2) Questions

Broad question (1):
What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?

Specific question:  
Can experience, gender, and played_hours predict whether a player subscribed to the newsletter?

How the data will help:  
- Compare subscription behaviour across experience groups
- Check whether gender influences subscription rates
- Explore whether players who spend more time playing are more likely to subscribe
- Later, we can use a logistic regression model to analyze whether experince, gender and played hours can predict subsription status
- Another thing that is helpful is that all the data for this specific question is within the players.csv dataset so we don't need to do any merging of datasets

```r
## (3) Exploratory Data Analysis and Visualization

library(tidyverse)
library(repr)
library(tidymodels)
library(cowplot)
options(repr.matrix.max.rows = 6)
#source('cleanup.R')

# a) Loading the data into R
players = read_csv("data/players.csv")
players
sessions = read_csv("data/sessions.csv")
sessions 
```

```r
# b) Minimal wrangling
# Making sure that the variable types were the correct type
players <- players |>
  mutate(subscribe = as.logical(subscribe),
    played_hours = as.numeric(played_hours))
players

# c) Calculating mean values of the two numeric variables - played_hours and age
mean_table = players |>
  summarise(across(where(is.numeric), ~mean(.x, na.rm = TRUE)))
mean_table
#** table format???
```

```r
# d) Exploratory visualizations
ggplot(players, aes(x = experience, fill = subscribe)) +    # subscription rate by experince level
  geom_bar(position = "fill") +
  labs(title = "Subscription Rates Across Experience Levels",
       x = "Experience Level",
       y = "Proportion of Players",
        fill = "Subscribed?")
# shows which players are more likely to subscribe
# helps determine if experience is a useful predictor of subscription status 

ggplot(players, aes(x = subscribe, y = played_hours)) +    #played hours vs. subscription status
  geom_boxplot() +
  labs(title = "Played Hours by Subscription Status",
       x = "Subscribed?",
       y = "Played Hours")
# shows whether there is a trend between subcription status and the number of hours played
# helps determine whether played hours is a good predictor of subscription status 

ggplot(players, aes(x = gender, fill = subscribe)) +      #gender distribution among subscribers
  geom_bar(position = "fill") +
  labs(title = "Subscription Rates by Gender",
       x = "Gender",
       y = "Proportion of Players", 
        fill = "Subscribed?")
# helps determine whether there is a trend between subscription status and gender 
# helps determine whether gender is a good predictor of subscription status 
```

e) Summary of info gathered from the exploratory visualizations 
-
- Subscription behaviour differs across experience categories, suggesting experience may contribute to predicting newsletter signup.
- Players with higher played_hours appear more likely to subscribe, indicating that gameplay intensity may influence subscription behaviour.
- Gender differences in subscription rates are visible but not large, this variable may still offer some predictive power.
- These observations support the idea that logistic regression could help identify which player characteristics relate to newsletter subscription.

## (4) Methods and Plan 
For this project, I will use logistic regression to predict whether a player subscribed to the newsletter using experience, gender, and played_hours as the explanatory variables.

a) Why is the method (logistic regression) appropriate?
-
- The response variable subscribe is binary (TRUE/FALSE), which works great in logisitic regression 
- Logistic regression outputs estimated probabilities, making it easy to interpret the likelihood of subscription based on player characteristics
- It can handle both categorical variables (gender, experience) and numeric variables (played_hours), which are the explanatory variables I chose to use

b) Which assumptions are required, if any, to apply the method selected?
- 
- Independence of observations
    - Each row in players.csv represents a different player, so this assumption is reasonable.

- Linear relationship 
    - The relationship between played_hours and log-odds of subscription should be roughly linear.

- Enough observations in each category so that outlier effect remains small 
    - Gender and experience levels appear to have enough counts to compensate for any outliers that may skew the data.
 
- No multicollinearity
    - Experience, gender, and played_hours are conceptually different variables so there should be no issues regarding very linearly related predictors

c) What are the potential limitations or weaknesses of the method selected?
- 
- Unbalanced classes
    - If many more players subscribed or did not subscribe, the model may struggle to classify the minority class.

- Missing values
    - Some players have missing gender or experience values, which may require dropping those rows or applying simple imputation.

- Limited predictors
    - Since the question only uses players.csv, important behavioural data from sessions.csv is not included. This may limit predictive accuracy.

- Categorical variable complexity
    - Experience has multiple levels. Some levels may have small sample sizes, which can introduce noise.
